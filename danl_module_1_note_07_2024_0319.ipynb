# Reshaping DataFrames

```python
import pandas as pd
# Below is for an interactive display of DataFrame in Colab
from google.colab import data_table
data_table.enable_dataframe_formatter()
```

```python
df_wide = pd.DataFrame({
    'Weekday': ['Tuesday', 'Wednesday'],
    'Miami': [80, 83],
    'Rochester': [57, 62],
    'St. Louis': [71, 75]
})

df_long = pd.DataFrame({
    'Weekday': ['Tuesday', 'Wednesday', 'Tuesday', 'Wednesday', 'Tuesday', 'Wednesday'],
    'City': ['Miami', 'Miami', 'Rochester', 'Rochester', 'St. Louis', 'St. Louis'],
    'Temperature': [80, 83, 57, 62, 71, 75]
})
```

```python
df_wide
```

```python
df_long
```

```python
df_wide_to_long = (
    df_wide
    .melt()
)

df_wide_to_long
```

```python
df_wide_to_long = (
    df_wide
    .melt(
        id_vars= ['Weekday']
    )
)

df_wide_to_long
```

```python
df_wide_to_long = (
    df_wide
    .melt(
        id_vars= ['Weekday'],
        var_name = 'City',
        value_name = 'Temperature'
    )
)

df_wide_to_long
```

```python
df_long
```

```python
df_wide
```

```python
(
    df_wide
    .melt(
        id_vars = 'Weekday',
        var_name = 'City',
        value_name = 'Temperature',
        value_vars = ['Miami', 'Rochester']
    )
)
```

```python
df_long_to_wide = (
    df_long
    .pivot(
        index = 'Weekday',
        columns = 'City',
        values = 'Temperature'
    )
    .reset_index()
)

df_long_to_wide
```

# Joining DataFrames

```python
x = pd.DataFrame({
    'key': [1, 2, 3],
    'val_x': ['x1', 'x2', 'x3']
})

y = pd.DataFrame({
    'key': [1, 2, 4],
    'val_y': ['y1', 'y2', 'y3']
})

x
```

```python
y
```

```python
merge_inner = x.merge(y, on = 'key', how = 'inner')
merge_inner
```

```python
merge_inner = pd.merge(x, y, on = 'key', how = 'inner')
merge_inner
```

```python
merge_left = x.merge(y, on = 'key', how = 'left')
merge_left
```

```python
merge_outer = x.merge(y, on = 'key', how = 'outer')
merge_outer
```

```python
x = pd.DataFrame({
    'key':[1, 2, 2, 3],
    'val_x':['x1', 'x2', 'x3', 'x4']})
y = pd.DataFrame({
    'key':[1, 2],
    'val_y':['y1', 'y2'] })
x
```

```python
y
```

```python
one_to_many = x.merge(y, on='key',
                         how='left')
one_to_many
```

```python
x = pd.DataFrame({
  'key':[1, 2, 2, 3],
  'val_x':['x1','x2','x3','x4']})

y = pd.DataFrame({
  'key': [1, 2, 2, 3],
  'val_y': ['y1', 'y2', 'y3', 'y4'] })

x
```

```python
y
```

```python
many_to_many = x.merge(y, on='key',
                          how='left')

many_to_many
```

```python
x = pd.DataFrame({
  'key_x': [1, 2, 3],
  'val_x': ['x1', 'x2', 'x3']
})
x
```

```python
y = pd.DataFrame({
  'key_y': [1, 2],
  'val_y': ['y1', 'y2'] })
y
```

```python
keys_xy = x.merge(y, left_on = 'key_x',
                     right_on = 'key_y',
                     how = 'left')
keys_xy
```

## Q4
Identify the full name of the airline with the highest average dep_delay, considering only positive delays.



```python
flights_airlines = (
    flights
    .query('dep_delay > 0')
    .merge(airlines, on = 'carrier', how = 'left')
    .groupby(['name'])[['dep_delay']]
    .mean()
    .sort_values('dep_delay', ascending = False)
)

flights_airlines
```
